```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.sample(5)
```

```python
credits.sample(5)
```

```python
movies=movies.merge(credits,on='title')
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

mv=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
mv.sample(5)
```

```python
mv.isnull().sum()
```

```python
mv.dropna(inplace=True)
```

```python
mv.isnull().sum()
```

```python
mv.duplicated().sum()
```

```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
mv['genres']=mv['genres'].apply(convert)
```

```python
mv['keywords']=mv['keywords'].apply(convert)
```

```python
def convert2(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            l.append(i['name'])
            count+=1
        else:
            break
    return l
```

```python
mv['cast']=mv['cast'].apply(convert2)
```

```python
mv.sample(5)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
mv['crew']=mv['crew'].apply(fetch_director)
```

```python
mv.sample(5)
```

```python
mv['overview'] = mv['overview'].apply(lambda x:x.split())
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
mv['cast'] = mv['cast'].apply(collapse)
mv['crew'] = mv['crew'].apply(collapse)
mv['genres'] = mv['genres'].apply(collapse)
mv['keywords'] = mv['keywords'].apply(collapse)
```

```python
mv.head()
```

```python
mv['tags'] = mv['overview'] + mv['genres'] + mv['keywords'] + mv['cast'] + mv['crew']
```

```python
new = mv.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: x.lower())
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('John Carter')
```
